```python
import matplotlib.pyplot as plt 

# Data
phases = ['Powerplay (1-6)', 'Middle (7-15)', 'Death (16-20)']
avg_runs_per_ball = [1.237, 1.297, 1.619]

# Plot
plt.figure()
plt.bar(phases, avg_runs_per_ball)

# Labels and title
plt.xlabel('Match Phases')
plt.ylabel('Average Runs per Ball')
plt.title('Phase-wise Scoring Rate (IPL)')

# Show values on top
for i, value in enumerate(avg_runs_per_ball):
    plt.text(i, value + 0.02, str(value), ha='center')

plt.show()
```
